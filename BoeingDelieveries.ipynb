```python
import pandas as pd
import altair as alt
```

```python
from google.colab import files
uploaded = files.upload()
```

First, we need to upload the dataset.

```python
import io
deliveries = pd.read_csv(io.BytesIO(uploaded['OrdersandDeliveries.csv']))
deliveries.head()
```

Next, I worked on my preliminary visualization by selecting the 787 program aircraft to work on a small scale visualization. I then cleaned the data.

```python
dreamliner = ['787-10', '787-9', '787-8']
dreamliner_deliveries = deliveries[deliveries['Model Series'].isin(dreamliner)]
# dreamliner_deliveries = dreamliner_deliveries[dreamliner_deliveries['Delivery Year '] == '2012']
dreamliner_deliveries.head()
```

```python
dreamliner_deliveries['Delivery Total'] = pd.to_numeric(dreamliner_deliveries['Delivery Total'])
dreamliner_deliveries['Order Total'] = pd.to_numeric(dreamliner_deliveries['Order Total'])
dreamliner_deliveries['Delivery Year '] = pd.to_datetime(dreamliner_deliveries['Delivery Year '])
dreamliner_deliveries
```

This is the preliminary chart.

```python
alt.Chart(dreamliner_deliveries).mark_area().encode(
    x='Delivery Year ',
    y='sum(Delivery Total)',
    color = 'Model Series'
)
```

This is a preliminary iteration of the chart by adding a tooltip and changing it to a stacked bar chart.

```python
selection = alt.selection(type='single', fields = ['Model Series'])
# selection = alt.selection(type='multi', fields=['Region'])
alt.Chart(dreamliner_deliveries).mark_bar(size = 25).encode(
    x='Delivery Year ',
    y='sum(Delivery Total)',
    color = 'Model Series',
    tooltip = ['Model Series', 'sum(Delivery Total)'],
    opacity=alt.condition(selection,alt.value(1),alt.value(.2))
).add_selection(selection)
```

I cleaned the original data and added another column that will be used for a dropdown menu in subsequent cells.

```python
deliveries['Model'] = deliveries['Model Series'].str.split('-').str[0]
deliveries.drop(deliveries.tail(1).index,inplace=True)
deliveries
```

```python
deliveries['Delivery Total'] = pd.to_numeric(deliveries['Delivery Total'])
deliveries['Order Total'] = pd.to_numeric(deliveries['Order Total'])
deliveries['Delivery Year '] = pd.to_datetime(deliveries['Delivery Year '])
```

```python
models = ['787','777','767']
deliveries = deliveries[deliveries['Model'].isin(models)]
deliveries
```

Here is the final product with dropdowns, tooltips, and highlighed stacked bar charts when you click on the model series on the legend.

```python
input_dropdown = alt.binding_select(options=['767','777','787'], name='Model')
selection = alt.selection_single(fields=['Model'], bind=input_dropdown)
click = alt.selection_single(fields = ['Model Series'], bind = 'legend')
alt.Chart(deliveries).mark_bar().encode(
    x='Delivery Year ',
    y='sum(Delivery Total)',
    color = 'Model Series',
    tooltip = ['Model Series', 'sum(Delivery Total)'],
    opacity=alt.condition(click,alt.value(1),alt.value(.2))
).add_selection(
    selection,
    click
).transform_filter(
    selection
)
```
